```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
fig=plt.figure(figsize=(8,4.5))
ax1=fig.add_subplot(1,2,1)
ax1.plot(x,y,linestyle="-.",linewidth=2,marker='o',markersize=0.2)
ax1.set_title("dshadja")
ax1.set_xlim([0,5])
ax1.set_ylim([0,1])
ax2=fig.add_subplot(2,2,2)
ax2.plot(x,y)
ax3=fig.add_subplot(2,2,4)
ax3.plot(x,y)
plt.show()
```
